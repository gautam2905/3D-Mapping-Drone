```python
# os.system("cd yolodata")
os.mkdir("yolodata")
for img_index, r in enumerate(result):
    img = r.orig_img
    box = r.boxes
    cls = box.cls
    for i,cl in enumerate(cls):
        cls_int = int(cl.item())  # Convert tensor to integer
        cls_name = class_dic[cls_int]
        if(os.path.exists(f"yolodata/{cls_name}")):
            x1, y1, x2, y2 = int(box.xyxy[i][0]), int(box.xyxy[i][1]), int(box.xyxy[i][2]), int(box.xyxy[i][3])
            x1, y1 = max(0, x1-30), max(0, y1-30)
            x2, y2 = min(img.shape[1], x2+30), min(img.shape[0], y2+30)
            if x1 < x2 and y1 < y2:
                crop_image = img[y1:y2, x1:x2, :]
                cv2.imwrite(f"yolodata/{cls_name}/{img_index}_{i}.jpg", crop_image)  
        else:
            os.mkdir(f"yolodata/{cls_name}")
            x1, y1, x2, y2 = int(box.xyxy[i][0]), int(box.xyxy[i][1]), int(box.xyxy[i][2]), int(box.xyxy[i][3])
            x1, y1 = max(0, x1-30), max(0, y1-30)
            x2, y2 = min(img.shape[1], x2+30), min(img.shape[0], y2+30)
            if x1 < x2 and y1 < y2:
                crop_image = img[y1:y2, x1:x2, :]
                cv2.imwrite(f"yolodata/{cls_name}/{img_index}_{i}.jpg", crop_image)
```
